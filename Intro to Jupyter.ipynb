````python
import requests
url = "https://pokeapi.co/api/v2/type/electric/"
response = requests.get(url)
data = response.json()
````

````python
data
````

````python
data['name']
````

````python
data.keys()
````

````python
data['names']
````

````python
data['names'][1].keys()
````

````python
for name in data['names']:
    print(name)
    if name['language']['name'] == "ko":
        print("This is the Korean word for electric type pokemon.")
        print(name['name'])
    print('----------')
````

````python
# Pokemon Speed (Pikachu vs Eevee)
````

````python
pika_url = "https://pokeapi.co/api/v2/pokemon/pikachu/"
response = requests.get(pika_url)
data = response.json()
````

````python
data
````

````python
data['stats'][-1]
````

````python
for stat in data['stats']:
    print(stat['stat']['name'])
    print('-------')
````

````python
for stat in data['stats']:
    print("-----")
    if stat['stat']['name'] == 'speed':
        print("FOUND SPEED!!!")
        # print(stat)
        print(stat['base_stat'])
        pikachu_speed = stat['base_stat']
print("Pikachu's speed is", pikachu_speed)
````

````python
eevee_url = "https://pokeapi.co/api/v2/pokemon/eevee/"
response = requests.get(eevee_url)
data = response.json()
````

````python
for stat in data['stats']:
    print("-----")
    if stat['stat']['name'] == 'speed':
        print("FOUND SPEED!!!")
        # print(stat)
        print(stat['base_stat'])
        eevee_speed = stat['base_stat']
print("Eevee's speed is", eevee_speed)
````
